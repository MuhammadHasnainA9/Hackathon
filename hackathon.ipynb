```python
#question 6
import random

def load_questions(filename):
    with open(filename, "r") as f:
        lines = f.readlines()
    questions = [line.strip().split("|") for line in lines]
    return questions

questions = load_questions("question.txt")
selected = random.sample(questions, 5)
score = 0

for q in selected:
    print("Q:" + q[0])
    print("A)", q[1])
    print("B)", q[2])
    print("C)", q[3])

    ans = input("Your answer (A/B/C): ").upper()
    if ans == q[4]:
        score += 1

print("Final Score:", score)

with open("leaderboard.txt", "w") as f:
    f.write("Score: " + str(score))
```

```python
# question 7
class Vehicle:
    def __init__(self, vehicletype):
        self.vehicletype = vehicletype

class ParkingLot:
    def __init__(self):
        self.slots = [None] * 5   

    def park_vehicle(self, vehicle):
        for i in range(len(self.slots)):
            if self.slots[i] is None:
                self.slots[i] = vehicle
                print(vehicle.vehicletype, "parked at slot", i)
                return
        print("Error: Parking Lot Full")

    def remove_vehicle(self, slot, hours):
        if self.slots[slot] is not None:
            self.slots[slot] = None
            print("Fee:", hours * 2)
        else:
            print("Slot already empty")

p = ParkingLot()

while True:
    print("1 Park Car")
    print("2 Park Bike")
    print("3 Remove Vehicle")
    print("4 Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        p.park_vehicle(Vehicle("Car"))

    elif choice == "2":
        p.park_vehicle(Vehicle("Bike"))

    elif choice == "3":
        s = int(input("Enter slot number: "))
        h = int(input("Enter hours parked: "))
        p.remove_vehicle(s, h)

    elif choice == "4":
        break
```

```python
# question 10
seats = [[0 for _ in range(5)] for _ in range(5)]

def show_seats():
    for row in seats:
        print(row)

while True:
    print("1 Book Seat")
    print("2 Show Seats")
    print("3 Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        r = int(input("Enter row (0-4): "))
        c = int(input("Enter column (0-4): "))

        if seats[r][c] == 0:
            seats[r][c] = 1
            print("Seat Booked")
        else:
            print("Seat Already Booked")

    elif choice == "2":
        show_seats()

    elif choice == "3":
        break
```

```python
# question 11
class Order:
    def __init__(self, order_id, item, vip):
        self.order_id = order_id
        self.item = item
        self.vip = vip

queue = []

while True:
    print("1 Add Normal Order")
    print("2 Add VIP Order")
    print("3 Process Next Order")
    print("4 Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        oid = input("Order ID: ")
        item = input("Item name: ")
        queue.append(Order(oid, item, False))
        print("Normal order added")

    elif choice == "2":
        oid = input("Order ID: ")
        item = input("Item name: ")
        queue.insert(0, Order(oid, item, True))
        print("VIP order added")

    elif choice == "3":
        if queue:
            o = queue.pop(0)
            print("Processing order:", o.item)
        else:
            print("No orders in queue")

    elif choice == "4":
        break
```
